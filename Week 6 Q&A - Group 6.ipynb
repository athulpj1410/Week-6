```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("Dataset_w6.xlsx")

df.head()
```

```python
df.isna().sum()
```

```python
df.country.nunique()
```

```python
# There are 196 countries in the dataset
```

```python
plt.figure(figsize=(15,15))
l=df['Region'].value_counts()
plt.pie(l,labels=l.index,shadow=True, textprops={'fontsize': 14},autopct='%1.2f%%')
plt.show()
```

```python
# It shows the percentage of countries lying in various regions
```

```python
plt.figure()
df.groupby("Region")['country'].count().plot(kind='bar',legend=True)
plt.show()
```

```python
#it shouws the count of various countries region wise
```

```python
plt.figure(figsize=(16,6))
df.groupby("Region")['Agricultural production index '].mean().plot(kind = "bar",color="#32a852")
plt.title("Region v/s Agricultural production index ",fontsize = 16)
plt.xticks(rotation = 45,fontsize = 10)
plt.show()
```

```python
# In general there is no extreme differnce in Agricultural production index of various regions so all are
# performing well in agriculture sector. Considering numerics South eastern asia performs well and Micronesia
# is behind
```

```python
# Number of Countries in each region

df.groupby('Region')["country"].count()
```

```python
plt.figure(figsize=(10,6))
l=df.groupby('Region')['GDP per capita (current US$)'].mean().plot(kind='bar')
plt.title("Average GDP Per Capita - Region wise",fontsize=16)
plt.show()
```

```python
# On an average North america has high GDP per capita and it faced a tight competition from Western Europe
```

```python
plt.figure(figsize=(10,6))
l = df.groupby("Region")['International trade: Imports (million US$)'].mean().plot(kind='bar')
plt.title("Average International trade : Import- Region wise",fontsize = 16)
plt.show()
```

```python
# Northern America again bags the first place for best avg in International Trade : imports
```

```python
plt.figure(figsize = (10,6))
l=df.groupby('Region')['GDP per capita (current US$)'].std().plot(kind='bar')
plt.title("Consistent GDP Per Capita", fontsize = 16)
plt.show()
```

```python
# Northern Africa and Western africa has got less standard deviation so they have consistent gdp considering
# it as a region all the countries have a gdp which are closer in numeric
```
